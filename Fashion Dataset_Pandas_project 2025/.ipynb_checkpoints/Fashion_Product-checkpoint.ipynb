# Fashion Dataset: Pandas Project for Data Analyst 2025

#### Project Details:
In this project we are going to study the csv file which contain fashion products sales.
We have below details in the given *fashion_product.csv*:

    # UserID : Unique ID for user
    # Product ID : Unique ID for product
    # Product Name : Product Name (Dress, Shoes, T-Shirt,etc)
    # Brand : Different brands product like (Adidas, H&M, Gucci, Zara,etc.)
    # Category : whether it's for men's, Women's or for Kids
    # Price : Price are in dollar
    # Rating : is in float range between 1 to 5
    # Color : color of the items
    # Size : different size (S, M, L, XL)


### Let's get started with the Data Extraction. 
The source of the data is from the kaggele.com it's public dataset

```python
# Import the required library of python which we are going to use.
import pandas as pd
import numpy as np
import matplotlib as mplot
import matplotlib.pyplot as plt
```

```python
# Let's get the dataset into dataframe to preform and analyze the data.

df=pd.read_csv("fashion_products.csv")
df.head()
```

```python
df.info()
```

```python
#Let's set the index number from 1 onwards
df.index=pd.RangeIndex(start=1,stop=len(df)+1)
```

```python
# Checking for null values
df.isnull().sum()
```

```python
categorical_columns = ['Product Name','Brand','Category','Price','Rating']
for column in categorical_columns:
    print(f"\nValue counts for '{column}':")
    print(df[column].value_counts(dropna=False))
```

```python
    category_counts = df['Brand'].value_counts()
    plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=90)
    plt.title('Total Brand sales in Pie Chart') # Add a title to your chart
    plt.axis('equal') # Ensures the pie chart is drawn as a circle.
    plt.show()
```

```python
category_counts = df.groupby(df["Brand"])["Price"].sum()
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Total Sales By Brand sales in Pie Chart') # Add a title to your chart
plt.axis('equal') # Ensures the pie chart is drawn as a circle.
plt.show()
```

```python
# Prepare data for the bar chart
products = df['Product Name']
sales = df['Size']

# Create the bar chart
plt.bar(products, sales, color='skyblue') # Adding a color for visual appeal

# Add labels and title
plt.xlabel('Product Name')
plt.ylabel('Sales')
plt.title('Sales by Product')

# Display the chart
plt.show()
```
