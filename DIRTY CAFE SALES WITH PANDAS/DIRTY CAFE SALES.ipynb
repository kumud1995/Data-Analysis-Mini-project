```python
#let's make a price dict
item_price = {
    'Coffee':2.0,
    'Tea':1.5,
    'Sandwich':4.0,
    'Salad':3.0,
    'Cake':3.0,
    'Cookie':1.0,
    'Smoothie':4.0,
    'Juice':3.0
}

# Reverse mapping (price to item)
price_item = {price: item for item, price in item_price.items()}

# Filling missing 'item' values based on 'Price per Unit'
df['Item'] = df['Item'].fillna(df['Price Per Unit'].map(price_item))

df.describe()
```

```python
try:
    df['Price Per Unit'] = df['Price Per Unit'].astype(float)
except Exception as e:
    print(e)
    pass
```

```python
import numpy as np
df['Price Per Unit'] = df['Price Per Unit'].replace(['ERROR','UNKNOWN'], np.nan)
df['Price Per Unit'] = df['Price Per Unit'].astype(float)
df.info()
```

```python
df['Item'] = df['Item'].fillna(df['Price Per Unit'].map(price_item))
df.info()
```

```python
df['Total Spent'] = df['Total Spent'].replace(['ERROR','UNKNOWN'], np.nan)
df['Total Spent'] = df['Total Spent'].astype(float)
```

```python
df['Quantity'] = df['Quantity'].replace(['ERROR','UNKNOWN'], np.nan)
df['Quantity'] = df['Quantity'].astype(float)
```

```python
# Filling the null values in column Price Per Unit
df['Price Per Unit'] = df['Price Per Unit'].fillna(df['Total Spent']/df['Quantity'])
df.info()
```

```python
# Fill Item Columns
df['Item'] = df['Item'].fillna(df['Price Per Unit'].map(price_item))
df.info()
```

```python
# WE still have missing items and since we can't possible determine 
#what are these items we are going to drop them

df.dropna(subset=['Item'], inplace = True)
df.info()
```

```python
# We are going to removethe missing / null values from Transaction Date
df.dropna(subset = ['Transaction Date'], inplace = True)
df.info()
```

```python
#We can still see some missing values in the 'Price Per Unit' column 
#we can use the same approach we used on 
#the 'Item' column but using item_price dictionary this time

df['Price Per Unit'] = df['Price Per Unit'].fillna(df['Item'].map(item_price))
df.info()
```

```python
empty_price_df = df[df['Price Per Unit'].isna()]
empty_price_df
```

```python
df = df[~df['Item'].isin(['ERROR','UNKNOWN'])]
df.info()
```
